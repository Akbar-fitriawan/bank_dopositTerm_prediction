# **Bank-marketing** ðŸ¦


Bank-marketing biasanya mengacu pada strategi dan teknik yang digunakan bank untuk mempromosikan produk dan layanan mereka kepada pelanggan potensial. Ini dapat mencakup berbagai saluran pemasaran seperti iklan, email langsung, pemasaran digital, media sosial, dan banyak lagi. Tujuan dari pemasaran bank adalah untuk menarik pelanggan baru, mempertahankan yang sudah ada, dan pada akhirnya meningkatkan pendapatan untuk bank. Strategi dapat melibatkan menargetkan demografi tertentu, menawarkan promosi atau insentif, dan memberikan pengalaman pelanggan yang dipersonalisasi. Selain itu, dengan munculnya perbankan online, banyak bank lebih fokus pada upaya pemasaran digital untuk menjangkau pelanggan melalui situs web, aplikasi mobile, dan platform online lainnya.

#### **Dataset Informasi :**

Data ini berkaitan dengan campaign pemasaran langsung dari sebuah lembaga perbankan di Portugal. Campaign pemasaran ini dilakukan melalui panggilan telepon. Sering kali, lebih dari satu kontak ke klien yang sama diperlukan untuk menilai apakah produk (deposito berjangka bank) akan disubscribe ('yes') atau tidak ('no').

```python
# Buat feature group_age
def age_group(age):
  if age <= 30:
    return 'youth segment(17-30y)'
  elif age > 30 and age <= 45:
    return 'mid-career(31-45y)'
  elif age > 45 and age < 65:
    return 'pre-retirement(46-64y)'
  elif age >= 65:
    return 'retirees(65+)'
  else:
    return 'unknown'

df['age_group'] = df['age'].apply(age_group)

# drop age
df.drop('age', axis=1, inplace=True)
```

Interpretasi:


- **17-30**: Usia muda dan awal dewasa, seringkali fokus pada pendidikan dan awal karir.
- **31-45**: Profesional dan keluarga muda, biasanya memerlukan produk seperti hipotek dan tabungan.
- **46-65**: Usia menengah dan mendekati pensiun, berfokus pada perencanaan pensiun dan investasi jangka panjang.
- **66-98**: Pensiunan, berfokus pada pengelolaan dana pensiun dan perencanaan warisan.

### Hari terbaik apa untuk menghubungi customers berdasarkan day of week, tipe contact, dan durasi panggilan, dengan target "yes"?


**Tujuan analysis**
>Menentukan hari terbaik untuk menghubungi klien berdasarkan durasi panggilan yang lebih singkat pada kontak sukses (target "yes"), yang menunjukkan minat tinggi dari klien.

- Uji hypotesa perbedaan rata-rata durasi panggilan antara hari-hari dalam seminggu.
- Uji perbedaan durasi panggilan antara tipe kontak yang berbeda pada hari tertentu.




**Uji Rata-Rata Durasi per Hari (ANOVA)**
> Hipotesis Null (H0): Tidak ada perbedaan signifikan dalam durasi panggilan antara hari-hari dalam seminggu untuk target "yes."

> Hipotesis Alternatif (H1): Ada perbedaan signifikan dalam durasi panggilan antara hari-hari dalam seminggu untuk target "yes," menunjukkan bahwa beberapa hari lebih baik untuk menghubungi klien.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
# Filter data untuk target "yes" saja
df_yes = df[df['y'] == 'yes']

# Tentukan urutan hari yang benar
day_order = ['mon', 'tue', 'wed', 'thu', 'fri']

# Mengelompokkan data berdasarkan 'day_of_week' dan 'contact' untuk menghitung rata-rata 'duration'
grouped_yes = df_yes.groupby(['day_of_week', 'contact']).agg(
    average_duration=('duration', 'mean'),
    count_contact=('contact', 'size')
).reset_index()

# Uji ANOVA untuk durasi per hari
anova_results = {}
for day in df_yes['day_of_week'].unique():
    anova_results[day] = df_yes[df_yes['day_of_week'] == day]['duration']

# ANOVA Test
f_value, p_value = stats.f_oneway(*anova_results.values())
print('=== ANOVA Results ===')
print(f'F-value: {f_value:.4f}')
print(f'p-value: {p_value:.4f}')

# Interpreting ANOVA
if p_value < 0.05:
    print("Significant difference found between days of the week. Proceed with Post-Hoc analysis.")
else:
    print("No significant difference found between days of the week.")


# Model ANOVA
model = ols('duration ~ C(day_of_week)', data=df_yes).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print('\n=== Detailed ANOVA Table ===')
print(anova_table)

# Post-hoc test menggunakan Tukey's HSD untuk menentukan hari terbaik
tukey = pairwise_tukeyhsd(endog=df_yes['duration'], groups=df_yes['day_of_week'], alpha=0.05)
print('\n=== Tukey HSD Test Results ===')
print(tukey)

# Visualisasi Tukey HSD untuk menunjukkan perbandingan antara hari
tukey.plot_simultaneous(figsize=(10, 6))
plt.title('Tukey HSD Test for Day of Week')
plt.xlabel('Mean Difference in Duration')
plt.show()

# Uji t untuk perbedaan durasi antara metode kontak 'cellular' dan 'telephone' pada hari tertentu
for day in day_order:
    data_cellular = df_yes[(df_yes['day_of_week'] == day) & (df_yes['contact'] == 'cellular')]['duration']
    data_telephone = df_yes[(df_yes['day_of_week'] == day) & (df_yes['contact'] == 'telephone')]['duration']
    if len(data_cellular) > 1 and len(data_telephone) > 1:  # Pastikan cukup data untuk uji t
        t_stat, p_val = stats.ttest_ind(data_cellular, data_telephone, equal_var=False)
        print(f'\n=== T-Test for {day.capitalize()} ===')
        print(f'Test statistic: {t_stat:.4f}')
        print(f'p-value: {p_val:.4f}')
        if p_val < 0.05:
            print(f"Significant difference in duration between 'cellular' and 'telephone' on {day.capitalize()}.")
        else:
            print(f"No significant difference between contact methods on {day.capitalize()}.")
```

#### Kesimpulan Analisis Statistik untuk Hari Terbaik Menghubungi Klien

Berdasarkan hasil uji statistik yang dilakukan, berikut adalah kesimpulan yang dapat ditarik:

1. **Hasil ANOVA**:
   - **F-value: 1.5461, p-value: 0.1860**
   - **Interpretasi**: Tidak ada perbedaan signifikan dalam durasi panggilan rata-rata antara hari-hari dalam seminggu (`p-value > 0.05`). Ini berarti, berdasarkan durasi panggilan, tidak ada hari tertentu yang lebih baik dibandingkan hari lainnya dalam mencapai target "yes".
   - **Kesimpulan**: Semua hari memiliki peluang yang relatif sama untuk berhubungan dengan klien dengan durasi panggilan yang tidak berbeda secara signifikan.

2. **Hasil Tukey HSD Test**:
   - Tidak ada pasangan hari yang menunjukkan perbedaan signifikan dalam durasi panggilan (semua nilai p-adj > 0.05).
   - **Interpretasi**: Hari-hari dalam seminggu tidak memiliki durasi panggilan yang berbeda secara signifikan dalam hal mencapai target "yes".
   - **Kesimpulan**: Tidak ada hari tertentu yang lebih efektif dibandingkan yang lain berdasarkan durasi panggilan.

3. **Hasil Uji t antara 'cellular' dan 'telephone'**:
   - Uji t menunjukkan perbedaan durasi yang signifikan antara metode kontak `cellular` dan `telephone` pada setiap hari (`p-value < 0.05` untuk semua hari).
   - **Interpretasi**: Pada semua hari (Senin hingga Jumat), durasi panggilan `cellular` secara signifikan berbeda dari `telephone`. Biasanya, `cellular` lebih cepat dan cenderung lebih efektif dalam mencapai target "yes" dibandingkan `telephone`.
   - **Kesimpulan**: Metode kontak `cellular` lebih baik dibandingkan `telephone` karena durasi panggilan yang lebih singkat, yang menunjukkan minat klien yang lebih tinggi.

### Rekomendasi Berdasarkan Temuan:
- **Hari Panggilan**: Karena tidak ada perbedaan signifikan dalam durasi panggilan antar hari, panggilan bisa dilakukan pada hari apa saja dari Senin hingga Jumat.
- **Metode Kontak**: Fokus pada metode `cellular`, karena terbukti lebih efektif dengan durasi panggilan yang lebih singkat.
- **Pendekatan Berbasis Durasi**: Prioritaskan klien yang merespons dengan durasi panggilan yang lebih singkat saat menggunakan metode `cellular`, karena ini mengindikasikan minat tinggi.

### Kesimpulan Utama:
Durasi panggilan tidak berbeda signifikan antar hari, sehingga hari apa pun dapat digunakan untuk menghubungi klien. Namun, metode `cellular` menunjukkan performa yang lebih baik dibandingkan `telephone`, membuatnya menjadi pilihan yang lebih disarankan untuk mencapai target "yes".

## Cek cardinalty

Mengapa cek cardinality?, salah satu cara untuk mengatasi input buruk, efisiensi model, peningkatan interpretabilitas model dan sebagainya.

```python
def cek_unikValues(df):
    ''' Fungsi ini untuk mengecek nilai unik di setiap kolom kategorikal
    df = dataframe yang akan di cek
    return = dataframe dari hasil cek unik values
    '''
    cols_cat = df.select_dtypes(include=['object']).columns.to_list() # pilih kolom kategorikal
    data_list = []
    for column in df[cols_cat]:
      unique_values = df[column].unique()
      data_dict = {
          'nama_kolom':column,
          'nilai_unik':df[column].unique(),
          'count_unik':len(unique_values),
          'cardinality': 'high' if len(unique_values) > 10 else 'low '
      }
      data_list.append(data_dict)

    data = pd.DataFrame(data_list)

    return data

cek_unikValues(X_train)
```

```python
def boxplot_check_outliers(df, numerical_features):
    """
      function cek outlier dengan boxplot

      parameters:
      df (dataframe) : data yang akan di cek outlier
      numerical_features (list): list dari data numerik

    """
    # Calculate the number of rows needed
    num_plots = len(numerical_features)
    num_rows = (num_plots // 2) + (num_plots % 2)

    # Set up the figure
    fig, axes = plt.subplots(num_rows, 2, figsize=(15, 5 * num_rows))
    axes = axes.flatten()

    # Create a box plot for each numerical feature
    for i, feature in enumerate(numerical_features):
        sns.boxplot(y=df[feature], ax=axes[i])
        axes[i].set_title(f'Box plot of {feature}')
        axes[i].set_xlabel(feature)

    # Remove any empty subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

Seperti penjelasan di atas adanya outlier sehingga harus di atasi, dalam hal ini saya menggunakan metode Winsorizing.

Cara melakukannya adalah memfilter ditribusi normal dan tidak dengan menggunakan motode statistika skewnes atau kurtosis dan menset threshold.
- Threshold untuk Distribusi Normal:
Biasanya, nilai skewness yang mendekati nol (`misalnya, antara -0.5 hingga 0.5`) menunjukkan bahwa distribusi data cenderung normal.
- Threshold untuk Distribusi Skewed:
Distribusi data dikatakan skewed jika nilai skewnessnya melebihi nilai ambang tertentu. Secara umum, `nilai skewness di atas 0.5 atau di bawah -0.5` sering digunakan sebagai indikator bahwa distribusi data tidak simetris dan cenderung skewed.

```python
def handling_outlier_wisorized(X_train, X_test, X_val,cols_outlier):
  #cek data normal dan tidak normal
  list_cols_normal = []
  list_cols_Not_normal = []

  for col in X[cols_outlier]:

    skew = X_train[col].skew()
    kurtosis = X_train[col].kurt()

    if -0.5 < skew < 0.5 and kurtosis < 3:
      list_cols_normal.append(col)
      
    else:
      list_cols_Not_normal.append(col)

  # handling outlier
  winsorizer_normal_dist = Winsorizer(capping_method='gaussian',
                                      tail='both',
                                      fold=3,
                                      variables=list_cols_normal,
                                      missing_values='ignore')

  winsorizer_not_normal_dist = Winsorizer(capping_method='iqr',
                                 tail='both',
                                 fold=1.5,
                                 variables=list_cols_Not_normal,
                                 missing_values='ignore')

  X_train_capped = winsorizer_normal_dist.fit_transform(X_train)
  X_test_capped = winsorizer_normal_dist.transform(X_test)
  X_val_capped = winsorizer_normal_dist.transform(X_val)


  X_train_capped = winsorizer_not_normal_dist.fit_transform(X_train_capped)
  X_test_capped = winsorizer_not_normal_dist.transform(X_test_capped)
  X_val_capped = winsorizer_not_normal_dist.transform(X_val_capped)

  return X_train_capped, X_test_capped, X_val_capped
```

## Feature Selection


Feature selection adalah proses memilih subset fitur yang paling relevan dari keseluruhan fitur dalam dataset untuk membangun model machine learning. Tujuannya adalah untuk meningkatkan kinerja model, mengurangi kompleksitas, dan menghindari overfitting. Berikut adalah metode-metode umum untuk feature selection:

1. Metode Filter

Penjelasan: Menggunakan metode statistik untuk menilai relevansi setiap fitur secara individual terhadap target variabel.

Metode:
- Pearson Correlation: Mengukur kekuatan hubungan linier antara fitur dan target.
- Chi-Square Test: Mengukur independensi antara fitur kategorikal dan target kategorikal.
- ANOVA (Analysis of Variance): Mengukur perbedaan rata-rata antara fitur numerik dan target kategorikal.
- Mutual Information: Mengukur ketergantungan informasi antara fitur dan target.
2. Metode Wrapper

Penjelasan: Menggunakan model prediktif untuk mengevaluasi subset fitur berdasarkan kinerja model. Metode ini lebih terintegrasi dengan model tetapi dapat lebih mahal secara komputasi.

Metode:
- Forward Selection: Memulai dengan tidak ada fitur dan menambahkan fitur satu per satu yang meningkatkan kinerja model.
- Backward Elimination: Memulai dengan semua fitur dan menghapus fitur satu per satu yang paling sedikit berpengaruh pada kinerja model.
- Recursive Feature Elimination (RFE): Menggunakan model untuk mengidentifikasi fitur penting dan menghapus fitur yang kurang penting secara berulang.
3. Metode Embedded

Penjelasan: Mengintegrasikan feature selection sebagai bagian dari proses pelatihan model. Metode ini menilai fitur berdasarkan kontribusinya selama pelatihan model.

Metode:
- Lasso Regression: Menggunakan regularisasi L1 untuk mengurangi koefisien fitur yang kurang penting menjadi nol.
- Decision Trees: Mengukur pentingnya fitur berdasarkan pengurangan impurity (misalnya, Gini impurity atau entropy) dalam pohon keputusan.
- Random Forest Feature Importance: Menggunakan ensemble pohon keputusan untuk menilai pentingnya fitur berdasarkan seberapa sering fitur digunakan untuk membagi data.
 Pemilihan Metode

- **Filter Methods**: Baik untuk pre-processing dan lebih cepat, tetapi mungkin tidak mempertimbangkan interaksi fitur.
- **Wrapper Methods**: Lebih akurat dalam memilih fitur tetapi lebih mahal secara komputasi.
- **Embedded Methods**: Mengintegrasikan feature selection dengan pelatihan model dan dapat memberikan fitur yang relevan secara langsung.

Memilih metode yang tepat tergantung pada ukuran dataset, tujuan model, dan kompleksitas perhitungan yang dapat diterima.

```python
def correlation(dataset, threshold):
  '''
  fungsi mencari feature yang memiliki korelasi tinggi dengan setiap kolom untuk mendeteksi fitur redundan

  parameter:

  dataset: data yang akan di cek korelasi
  threshold: nilai threshold yang ditentukan
  return: kolom yang memiliki korelasi tinggi dengan target

  penggunaan : X = correlation(X_train_num, 0.7)
  '''
  col_corr = set()
  corr_matrix = dataset.corr()
  for i in range(len(corr_matrix.columns)):
    for j in range(i):
      if abs(corr_matrix.iloc[i,j]) > threshold:
        col_name = corr_matrix.columns[i]

        col_corr.add(col_name)
  return col_corr
```

```python
# fungsi evaluasi model Precision-Recall Curve dan ROC AUC.

def evaluate_models(models, X_test, y_test, thresholds=0.5):
    """
    Evaluates multiple models on the test data using Precision-Recall Curve and ROC AUC.

    Args:
        models (dict): A dictionary where keys are model names and values are model objects.
        X_test (array-like): Test features.
        y_test (array-like): True labels for the test set.

    Returns:
        None
    """
    plt.figure(figsize=(14, 6))

    # Plot Precision-Recall Curve
    plt.subplot(1, 2, 1)
    for name, model in models.items():
        # Predict probabilities
        y_scores = model.predict_proba(X_test)[:, 1]
        precision, recall, thresholds = precision_recall_curve(y_test, y_scores)
        pr_auc = average_precision_score(y_test, y_scores)

        # Plot Precision-Recall Curve
        plt.plot(recall, precision, label=f'{name} (AP: {pr_auc:.2f})')

    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend(loc='best')
    plt.grid(True)

    # Plot ROC Curve
    plt.subplot(1, 2, 2)
    for name, model in models.items():
        # Predict probabilities
        y_scores = model.predict_proba(X_test)[:, 1]
        fpr, tpr, thresholds = roc_curve(y_test, y_scores)
        roc_auc = roc_auc_score(y_test, y_scores)

        # Plot ROC Curve
        plt.plot(fpr, tpr, label=f'{name} (AUC: {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], linestyle='--', color='grey')  # Diagonal line for random guess
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc='best')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    # Print Precision, Recall, and ROC AUC scores
    for name, model in models.items():
        y_scores = model.predict_proba(X_test)[:, 1]
        pr_auc = average_precision_score(y_test, y_scores)
        roc_auc = roc_auc_score(y_test, y_scores)
        print("-"*50)
        print(f'{name} - Precision-Recall AUC: {pr_auc:.2f}, ROC AUC: {roc_auc:.2f}')
```

```python
import os
import joblib
import json
from datetime import datetime
from pprint import pprint
```

```python
# Utils untuk menyimpan model

# Direktori tempat menyimpan model dan parameter
PATH_DIR = './model_and_param'
os.makedirs(PATH_DIR, exist_ok=True)

def get_unique_filename(directory, base_filename, extension):
    """
    Menambahkan angka sebagai suffix jika file sudah ada.
    Contoh: jika 'model.pkl' ada, maka akan menjadi 'model_1.pkl', 'model_2.pkl', dst.

    Args:
        directory (str): Direktori tempat menyimpan file.
        base_filename (str): Nama dasar file tanpa suffix.
        extension (str): Ekstensi file ('pkl' atau 'json').

    Returns:
        str: Nama file unik dengan extension yang sesuai.
    """
    counter = 1
    filename = f"{base_filename}.{extension}"
    while os.path.exists(os.path.join(directory, filename)):
        filename = f"{base_filename}_{counter}.{extension}"
        counter += 1
    return filename

def save_model(model, model_name, search_type=None):
    """
    Menyimpan model dengan nama file unik.

    Args:
        model: Model yang akan disimpan.
        model_name (str): Nama model yang akan digunakan sebagai bagian dari nama file.
        search_type (str, optional): Jenis pencarian atau deskripsi tambahan untuk nama file.
    """
    # Membuat nama dasar file
    base_filename = f"{model_name.replace(' ', '_')}_model"
    if search_type:
        base_filename = f"{base_filename}_{search_type}"

    # Mendapatkan nama file unik untuk model
    filename = get_unique_filename(PATH_DIR, base_filename, 'pkl')
    joblib.dump(model, os.path.join(PATH_DIR, filename))
    print(f"Model {model_name} saved as {filename}")

def save_param(data, param_name):
    """
    Menyimpan parameter dalam format JSON dengan suffix numerik jika file sudah ada.

    Args:
        data: Data yang akan disimpan dalam file JSON.
        param_name (str): Nama dasar untuk file parameter yang akan disimpan.
    """
    base_filename = f"{param_name.replace(' ', '_')}_params"
    filename = get_unique_filename(PATH_DIR, base_filename, 'json')
    filepath = os.path.join(PATH_DIR, filename)

    # Menyimpan data dalam format JSON
    with open(filepath, 'w') as file:
        json.dump(data, file, default=int, indent=4)
    print(f"Parameters saved to {filename}")

def save_tuning_report(data, report_name):
    """
    Menyimpan hasil tuning dalam format JSON dengan timestamp untuk keunikan nama file.

    Args:
        data: Data hasil tuning yang akan disimpan.
        report_name (str): Nama dasar untuk file report yang akan disimpan.
    """
    timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
    filename = f"{report_name.replace(' ', '_')}_report_{timestamp}.json"
    filepath = os.path.join(PATH_DIR, filename)
    pd.DataFrame(data).to_json(filepath, orient='records')
    print(f"Report tuning saved to {filename}")

# Contoh penggunaan:
# save_model(model, 'log_reg', 'randomSearch')
# save_param(randomCV_params, 'randomCV')
# save_tuning_report(randomCV_results, 'randomCV_results')
```

```python
# Hyperparameter tuning random-search

def random_search_tuning_hyperparameter(models, X, y, param_distributions, n_iter=100, scoring='accuracy'):
    seed = 42
    np.random.seed(seed)

    best_models = {}
    best_params_list = []
    results_tuning = []

    # Stratified K-Fold Cross Validation
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)

    for model_name, model in models.items():
        # Mendapatkan distribusi parameter untuk model yang sedang dituning
        params = param_distributions.get(model_name, {})

        try:
            print(f"Starting Random Search for {model_name}...")

            # Inisialisasi pipeline dengan model yang spesifik
            model_pipe = ImbPipeline(steps=[
                ('preprocessor', preprocessor),  # Preprocessing step (seperti scaling)
                ('oversample', SMOTE(random_state=seed)),  # Oversampling step dengan SMOTE
                ('model', model)  # Model step
            ])

            # Inisialisasi RandomizedSearchCV
            random_search = RandomizedSearchCV(
                estimator=model_pipe,
                param_distributions=params,
                n_iter=n_iter,  # Jumlah iterasi yang ingin dicoba
                cv=skf,  # Cross-validation dengan StratifiedKFold
                scoring=scoring,  # Metric evaluasi, bisa diubah sesuai kebutuhan
                n_jobs=-1,  # Menggunakan semua core CPU yang tersedia
                verbose=2,
                random_state=seed
            )

            # Melakukan fitting RandomizedSearchCV
            random_search.fit(X, y)

            # Mendapatkan model terbaik, parameter terbaik, dan skor terbaik
            tuned_model = random_search.best_estimator_
            best_params = random_search.best_params_
            best_score = random_search.best_score_
            results = random_search.cv_results_

            print(f"{model_name} - Best Parameters:")
            pprint(best_params)
            print(f"Best Score: {best_score:.4f}")


            # Menyimpan model terbaik dengan parameter terbaik
            best_models[model_name] = tuned_model
            best_params_list.append({model_name : best_params})
            results_tuning.append(results)

            # save model
            save_model(tuned_model, model_name, search_type='randomSearch')
            # save param
            save_param(best_params, model_name)
            # save tuning report
            save_tuning_report(results, model_name)

            print('-' * 50, '\n')

        except Exception as e:
            print(f'Error Training {model_name}: {e}')

    return best_models, best_params_list, results_tuning
```

### Predict And Evaluate Model with data test

##### **1. Precision:**
- **Definisi:** Precision mengukur proporsi dari semua prediksi positif yang benar. Secara matematis, precision dapat dihitung dengan rumus:

$$
  \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}
$$

- **Relevansi:** Dalam konteks prediksi langganan produk deposito, precision menunjukkan seberapa akurat model ketika memprediksi bahwa seorang pelanggan akan berlangganan.
- Jika **precision** sangat penting (misalnya, untuk menghindari biaya pemasaran yang tinggi pada pelanggan yang tidak tertarik), maka fokus akan lebih pada memastikan bahwa setiap prediksi positif benar-benar relevan.

##### **2. Recall:**
- **Definisi:** Recall mengukur proporsi dari semua kasus positif yang benar-benar terdeteksi oleh model. Secara matematis, recall dapat dihitung dengan rumus:

$$
\text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}
$$

- **Relevansi:** Recall dalam konteks ini menunjukkan seberapa baik model dalam mengidentifikasi semua pelanggan yang benar-benar tertarik untuk berlangganan.

- Jika **recall** sangat penting (misalnya, untuk memastikan semua kemungkinan pelanggan diidentifikasi), maka model akan dirancang untuk menangkap sebanyak mungkin pelanggan potensial meskipun beberapa prediksi mungkin salah.

- **Konteks Bisnis:**
  - **Biaya False Positives:** Jika biaya menghubungi pelanggan yang tidak tertarik tinggi, maka precision menjadi metrik yang lebih penting untuk memastikan bahwa kontak yang dilakukan lebih efektif dan relevan.
  - **Biaya False Negatives:** Jika kehilangan pelanggan potensial lebih merugikan daripada menghubungi beberapa yang tidak tertarik, maka recall lebih penting untuk memastikan bahwa semua kemungkinan pelanggan terjangkau.

```python
def handling_outlier_wisorized(X_train, X_test, X_val, cols_outlier):
  #cek data normal dan tidak normal
  list_cols_normal = []
  list_cols_Not_normal = []

  for col in X[cols_outlier]:

    skew = X_train[col].skew()
    kurtosis = X_train[col].kurt()

    if -0.5< skew < 0.5 and kurtosis < 3:
      list_cols_normal.append(col)
    else:
      list_cols_Not_normal.append(col)

  # handling outlier
  winsorizer_normal_dist = Winsorizer(capping_method='gaussian',
                                      tail='both',
                                      fold=3,
                                      variables=list_cols_normal,
                                      missing_values='ignore')

  winsorizer_not_normal_dist = Winsorizer(capping_method='iqr',
                                 tail='both',
                                 fold=1.5,
                                 variables=list_cols_Not_normal,
                                 missing_values='ignore')

  X_train_capped = winsorizer_normal_dist.fit_transform(X_train)
  X_test_capped = winsorizer_normal_dist.transform(X_test)
  X_val_capped = winsorizer_normal_dist.transform(X_val)


  X_train_capped = winsorizer_not_normal_dist.fit_transform(X_train_capped)
  X_test_capped = winsorizer_not_normal_dist.transform(X_test_capped)
  X_val_capped = winsorizer_not_normal_dist.transform(X_val_capped)

  return X_train_capped, X_test_capped, X_val_capped
```

```python
# clear session
seed = 20
tf.keras.backend.clear_session()
np.random.seed(seed)
tf.random.set_seed(seed)

# define model
model_ann = Sequential([
  Dense(64, activation='relu', kernel_initializer=tf.keras.initializers.HeNormal(seed), input_shape=(X_train_resampled.shape[1:])), # input and hiden layer 1
  Dropout(0.2),
  Dense(32, activation='relu', kernel_initializer=tf.keras.initializers.HeNormal(seed)), # hiden layer 2
  Dropout(0.2),
  Dense(16, activation='relu', kernel_initializer=tf.keras.initializers.HeNormal(seed)), # hiden layer 3
  Dense(1, activation='sigmoid',  kernel_initializer=GlorotUniform(seed)) # ouput layer untuk binary classificatio
  ])

# Compile the model
model_ann.compile(optimizer='adam',
            loss='binary_crossentropy',  # For binary classification
            metrics=['accuracy'])

model_ann.summary()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc

def plot_training_history(histories, labels):
    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    for history, label in zip(histories, labels):
        plt.plot(history.history['accuracy'], label=f'{label} Train Accuracy')
        plt.plot(history.history['val_accuracy'], label=f'{label} Val Accuracy')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    for history, label in zip(histories, labels):
        plt.plot(history.history['loss'], label=f'{label} Train Loss')
        plt.plot(history.history['val_loss'], label=f'{label} Val Loss')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrices(y_true, y_pred_list, labels):
    plt.figure(figsize=(12, 8))
    for i, (y_pred, label) in enumerate(zip(y_pred_list, labels)):
        cm = confusion_matrix(y_true, y_pred)
        plt.subplot(1, len(y_pred_list), i+1)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(f'Confusion Matrix - {label}')
        plt.xlabel('Predicted')
        plt.ylabel('True')
    plt.tight_layout()
    plt.show()

def plot_roc_curves(y_true, y_probs_list, labels):
    plt.figure(figsize=(10, 8))
    for y_probs, label in zip(y_probs_list, labels):
        fpr, tpr, _ = roc_curve(y_true, y_probs)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{label} (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title('ROC Curve')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.show()

# Example usage:
# histories: list of history objects from Keras models
# labels: list of model names
# y_true: true labels
# y_pred_list: list of predicted labels
# y_probs_list: list of predicted probabilities

plot_training_history(result_df, model_ann)
# plot_confusion_matrices(y_val, y_pred_list, model_ann)
# plot_roc_curves(y_val, y_probs_list, labels)
```
